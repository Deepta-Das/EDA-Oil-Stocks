## Simple Moving Average(SMA)

SMAs are commonly used to smooth price data and technical indicators. The longer the period of the SMA, the smoother the result, but the more lag that is introduced between the SMA and the source.

```python
#Caculation of SMA for ONGC stock price
ONGC_df['Moving Average_9']=ONGC_df['Close'].rolling(window=9).mean()
ONGC_df['Moving Average_50']=ONGC_df['Close'].rolling(window=50).mean()
```

```python
#Plotting SMA for ONGC for stock price
sns.set_style("whitegrid")
plt.figure(figsize=(14,8))
plt.plot(ONGC_df.Date,ONGC_df.Close,c='darkred')
plt.plot(ONGC_df.Date,ONGC_df['Moving Average_9'])
plt.plot(ONGC_df.Date,ONGC_df['Moving Average_50'])
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(['Close','Moving_average_9','Moving_average_50'])
plt.title(":Simple Moving Average Plot:ONGC Closing Price Vs Date");
```

```python
#Calculation of SMA for IOC stock price
IOC_df['Moving Average_9']=IOC_df['Close'].rolling(window=9).mean()
IOC_df['Moving Average_50']=IOC_df['Close'].rolling(window=50).mean()
```

```python
#Plotting SMA for IOC for stock price
sns.set_style("darkgrid")
plt.figure(figsize=(14,8))
plt.plot(IOC_df.Date,IOC_df.Close,c='orangered')
plt.plot(IOC_df.Date,IOC_df['Moving Average_9'])
plt.plot(IOC_df.Date,IOC_df['Moving Average_50'])
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(['Close','Moving_average_9','Moving_average_50'])
plt.title(":Simple Moving Average Plot:IOC Closing Price Vs Date");
```

```python
#Caculation of SMA for BPCL stock price
BPCL_df['Moving Average_9']=BPCL_df['Close'].rolling(window=9).mean()
BPCL_df['Moving Average_50']=BPCL_df['Close'].rolling(window=50).mean()
```

```python
#Plotting SMA for BPCL for stock price
sns.set_style("whitegrid")
plt.figure(figsize=(14,8))
plt.plot(BPCL_df.Date,BPCL_df.Close,c='b')
plt.plot(BPCL_df.Date,BPCL_df['Moving Average_9'])
plt.plot(BPCL_df.Date,BPCL_df['Moving Average_50'])
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(['Close','Moving_average_9','Moving_average_50'])
plt.title("Simple Moving Average Plot:BPCL Closing Price Vs Date");
```

## Daily Return 

Daily return on a stock is used to measure the day to day performance of stocks, it is the price of stocks at todayâ€™s closure compared to the price of the same stock at yesterdayâ€™s closure.  A positive return means the stock has grown in value, while a negative return means it has lost value.

```python
#Calculating Daily Returns of each company
ONGC_df['Daily_Returns']=ONGC_df.Close.pct_change()*100
IOC_df['Daily_Returns']=IOC_df.Close.pct_change()*100
BPCL_df['Daily_Returns']=BPCL_df.Close.pct_change()*100
```

```python
#Plotting Daily Return of the stocks in one output, by having subplots 
fig, axes = plt.subplots(3, 1, figsize=(14,17))

# Using the axes for plotting
axes[0].stem(ONGC_df.Date,ONGC_df.Daily_Returns,'DarkRed',use_line_collection=True)
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Daily Returns')
axes[0].set_title('Daily Returns of ONGC stock')

axes[1].stem(IOC_df.Date,IOC_df.Daily_Returns,'OrangeRed',use_line_collection=True )
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Daily Returns')
axes[1].set_title('Daily Returns of IOC stock')

axes[2].stem(BPCL_df.Date,BPCL_df.Daily_Returns,'b',use_line_collection=True )
axes[2].set_xlabel('Date')
axes[2].set_ylabel('Daily Returns')
axes[2].set_title('Daily Returns of BPCL stock')
plt.tight_layout(pad=2);
```

The above stem plots(lollipop plots) give the daily returns of each company. If the vertical lines are above the red baseline it means a positive return and if the vertical lines are below the red baseline it means a negative return on that day.

# Step 4: Ask & answer questions about the data

## Q1)Which stock is the most volatile?

```python
sns.set_style("whitegrid")
plt.figure(figsize=(14,8))
plt.plot(ONGC_df.Date,ONGC_df.Daily_Returns.rolling(30).std(skipna = True),c='darkred')
plt.plot(IOC_df.Date,IOC_df.Daily_Returns.rolling(30).std(skipna = True),c='orangered')
plt.plot(BPCL_df.Date,BPCL_df.Daily_Returns.rolling(30).std(skipna = True),c='b')
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.title("Oil Companies Volatility")
plt.legend(["ONGC","IOC","BPCL"]);
```

The above plot show the comparative volatility analysis among the three oil companies,viz ONGC, IOC and BPCL.We plot this by computing the 30-day(approx.1 month) rolling mean(also called moving average) of the daily returns, then compute the standard deviation and plot the values.

*The above plot show that BPCL stock is the most volatile for most part of the  given timeframe.*

## Q4)What was the effect of COVID19 on the price of the stocks

```python
#Changing dtype of Date to datetime64 from object
ONGC_df['Date']=pd.to_datetime(ONGC_df['Date'])
IOC_df['Date']=pd.to_datetime(IOC_df['Date'])
BPCL_df['Date']=pd.to_datetime(BPCL_df['Date'])
```

```python
#Changing the index to Date for slicing
ONGC_df= ONGC_df.set_index('Date')
IOC_df= IOC_df.set_index('Date')
BPCL_df= BPCL_df.set_index('Date')
```

```python
#Generating subplots
fig, axes = plt.subplots(3, 1, figsize=(14,17))
plt.figtext(.14, .90, "*COVID 19 was declared as a pandemic on 11 March 2020", c='red')
#Plot for ONGC stock price 
ONGC_df_corona = ONGC_df['2020-03-11':]
axes[0].plot(ONGC_df[:'2020-03-11'].Open,'g')
axes[0].plot(ONGC_df_corona.Open,'r')
axes[0].legend(['Pre Pandemic', 'During Pandemic'])
axes[0].set_ylabel('Stock opening price');
axes[0].set_title('ONGC stock prices during COVID 19 pandemic')
#Plot for IOC stock price 
IOC_df_corona = IOC_df['2020-03-11':]
axes[1].plot(IOC_df[:'2020-03-11'].Open,'g')
axes[1].plot(IOC_df_corona.Open,'r')
axes[1].legend(['Pre Pandemic', 'During Pandemic'])
axes[1].set_ylabel('Stock opening price');
axes[1].set_title('IOC stock prices during COVID 19 pandemic')
#Plot for BPCL stock price 
BPCL_df_corona = BPCL_df['2020-03-11':]
axes[2].plot(BPCL_df[:'2020-03-11'].Open,'g')
axes[2].plot(BPCL_df_corona.Open,'r')
axes[2].legend(['Pre Pandemic', 'During Pandemic'])
axes[2].set_ylabel('Stock opening price');
axes[2].set_title('BPCL stock prices during COVID 19 pandemic');
```

From the plots, we can see that the COVID19 pandemic had a huge impact on the stock prices of ONGC, IOC and BPCL stocks. The day COVID19 was declared a pandemic , all the stock prices started falling. ONGC and IOC could not regain their pre covid stock prices,but BPCL stock price managed to go up after an initial low.
